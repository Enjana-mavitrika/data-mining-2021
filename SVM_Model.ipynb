```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
#Loading the dataset :

df_users_labels_test = pd.read_csv("Data/users/coded_ids_labels_test.csv")
df_users_ids = pd.read_csv("Data/users/coded_ids.csv")
df_users_labels_train = pd.read_csv("Data/users/coded_ids_labels_train.csv")
df_features = pd.read_csv("Data/users_features/features.csv")

df_users= pd.merge(df_users_ids,df_users_labels_train,how="right")
userfeaters= pd.merge(df_users,df_features,how="left")

userfeaters
```

```python
userfeaters.drop(['utc_offset', 'urls_used_on_average', 'followers_count_minus_2002', 'time_zone' ,'lang', 'date_newest_tweet', 'date_oldest_tweet', 'max_intertweet_times', 'min_intertweet_times', 'std_intertweet_times', 'avg_intertweet_times'], inplace=True, axis=1)
```

```python
userfeaters['default_profile']=userfeaters['default_profile'].replace({True :1, False: 0})
userfeaters['default_profile_image']=userfeaters['default_profile_image'].replace({True :1, False: 0})
userfeaters
```

```python
df=userfeaters.dropna(axis=1, how='all')
```

```python
x=df
y=df["label"]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
```

```python
svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
